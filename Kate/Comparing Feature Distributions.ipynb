### Generate Histograms Function and Mann-Whitney U Test
#### Needed for numerical columns

```python
def generate_distribution_histogram(dataframe, column_name, number_bins, 
                                    alpha, color, edgecolor, label_name,
                                    density = True,
                                    title='', x_axis_label='', y_axis_label=''):
    
    plt.hist(dataframe[column_name], bins = number_bins, alpha = alpha, color = color, 
             edgecolor = edgecolor, label = label_name, density = density)
    plt.title(title)
    plt.xlabel(x_axis_label)
    plt.ylabel(y_axis_label)
    plt.legend(loc='upper left')
    

def mann_whitney_u_test(distribution_1, distribution_2):

    u_statistic, p_value = stats.mannwhitneyu(distribution_1, distribution_2)
    
    #Print the results
    print('U-Statistic: ', u_statistic)
    print('p-value: ', p_value)
    print()
```

### Split dataframes into training and testing

```python
# Split df1 into X_train and y_train
X_train1 = df1.iloc[:, :-1]
y_train1 = df1.iloc[:, -1]

# Split df2 into X_train and y_train
X_train2 = df2.iloc[:, :-1]
y_train2 = df2.iloc[:, -1]
```

#### Finding the number of columns in the datasets
##### (only have to do one because they should be the same)

```python
# Ensure data is 2D
if X_train1.ndim == 1:
    X_train1 = X_train1.reshape(-1, 1)  # Reshape 1D array to 2D array with one column

# Number of features (columns) in the dataset
X_train_num_features1 = X_train1.shape[1]
```

## Running through each column of the datasets to find either histogram distributions (numerical columns) or categorical probability (categorical columns) differences

```python
for column in range(X_train_num_features1):

    if X_train1[column].dtype == 'float64':

        # Calculate common bins based on the range of values in both datasets for this column
        combined_min = min(np.min(X_train1[column]), np.min(X_train2[column]))
        combined_max = max(np.max(X_train1[column]), np.max(X_train2[column]))
        bins = np.linspace(combined_min, combined_max, 10)  # Adjust the number of bins as needed
    
        generate_distribution_histogram(dataframe = X_train1, 
                                        column_name = column, 
                                        number_bins = bins, 
                                        alpha = 0.8, 
                                        color = 'teal', 
                                        edgecolor = 'black',
                                        label_name = 'Original',
                                        density = True,
                                        title=f'Column {column}',
                                        x_axis_label = 'Values', 
                                        y_axis_label = 'Frequencies')
        
        generate_distribution_histogram(dataframe = X_train2, 
                                        column_name = column, 
                                        number_bins = bins, 
                                        alpha = 0.3, 
                                        color = 'darkviolet', 
                                        edgecolor = 'black',
                                        label_name = 'Orig + New',
                                        density = True,
                                        title=f'Column {column}',
                                        x_axis_label = 'Values', 
                                        y_axis_label = 'Frequencies')

        plt.show()

        print(f"Column {column}:")
        mann_whitney_u_test(list(X_train1[X_train1.columns[column]]), list(X_train2[X_train2.columns[column]]))

    elif X_train1[column].dtype != 'float64':

        value_counts1 = X_train1.iloc[:, column].value_counts(normalize=True).sort_index()
        value_counts2 = X_train2.iloc[:, column].value_counts(normalize=True).sort_index()

        # Align indices
        all_categories = value_counts1.index.union(value_counts2.index)
        value_counts1 = value_counts1.reindex(all_categories, fill_value=0)
        
        print(f"Column {column}:")
        print("df1:", value_counts1.values)  # Print just the values for X_train1
        print("df2:", value_counts2.values)  # Print just the values for X_train2
        
        # Compute the L2 norm (Euclidean norm) of the difference between the arrays
        L2_norm = np.linalg.norm(value_counts1.values - value_counts2.values, ord=2)
        L2_norm = L2_norm / np.linalg.norm(value_counts1.values, ord=2)
        print("Euclidean norm of difference:", L2_norm)  # Print the differences
        print()
```
